```python
# Get the data for the stock you want to calculate PNL for
data = yf.download("AAPL", start="2023-01-01", end="2023-03-08")

# Calculate the PNL
pnl = data["Adj Close"].pct_change() * data["Close"]

# Print the PNL
print(pnl)
```

```python
# Define the tickers
tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'META']

# Get data for the tickers
data = yf.download(tickers, start='2022-01-01', end='2023-07-15')['Adj Close']

print(data)
```

```python
data2 = yf.download(tickers, start='2022-01-01', end='2023-07-15')['Average']

print(data2)
```

```python
data2
```

```python
# Compute the purchase price and current price
purchase_price = data.iloc[0]
current_price = data.iloc[-1]

# Compute the profit and loss
pnl = current_price - purchase_price

# sort the values higest to lowest 
pnl = pnl.sort_values(ascending=False)

print(pnl)
print("******")

pnl_positive = pnl >0 

print(pnl_positive)
```

```python


# Visualize the profit and loss
plt.figure(figsize=(10, 6))
pnl.plot(kind='bar', color=pnl_positive.map({True: 'g', False: 'r'}))
plt.xticks(rotation=45)
plt.xlabel('Ticker')
plt.ylabel('Profit and Loss')
plt.title('Basic Profit and Loss')
plt.grid(True)
plt.show()
```

```python
# Visualize the profit and loss
plt.figure(figsize=(10, 6))
pnl.plot(kind='line')
plt.xticks(rotation=45)
plt.xlabel('Ticker')
plt.ylabel('Profit and Loss')
plt.title('Basic Profit and Loss')
plt.grid(True)
plt.show()
```
